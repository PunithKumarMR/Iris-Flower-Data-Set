# 02. Write a note about the data set
*In a markdown cell at the start of your notebook, write a short description of the iris data set, complete with references.*  

The Iris flower data set was introduced by the British statistician and biologist [Ronald Fisher](https://en.wikipedia.org/wiki/Ronald_Fisher) in 1936 as an example of [linear discriminant analysis](https://en.wikipedia.org/wiki/Linear_discriminant_analysis). The data set contains 50 samples from each of three species of Iris(Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals in centimeters. Based on the combination of these four features, Fisher developed a linear discriminant model to distinguish the species from each other.

# 06. Fit a line  
*Fit a straight line to the variables petal length and petal width for the whole data set. Plot the data points in a scatter plot with the best fit line shown.*  
Straight line:  
$$y = mx + c$$  

Calculating cost:  
$$ Cost(m, c) = \sum_i (y_i - mx_i - c)^2 $$

```python
#calculating the best values for m and c
w = petalLength
d = petalWidth

w_avg = np.mean(w)
d_avg = np.mean(d)

w_zero = w - w_avg
d_zero = d - d_avg

m = np.sum(w_zero * d_zero) / np.sum(w_zero * w_zero)
c = d_avg - m * w_avg

#setting figure size, title and labels
pl.rcParams['figure.figsize'] = (16, 8)
pl.title("Straight line using petal length & width", fontsize=20)
pl.xlabel("Petal Length", fontsize = 12)
pl.ylabel("Petal Width", fontsize = 12)

#plotting the data and the line
pl.plot(w, d, 'x', label='Iris Data')
pl.plot(w, m * w + c, 'g', label='Best Fit Line')
#displaying the legend and the diagram
pl.legend()
pl.show()
```

# 07. Calculate the R-squared value  
*Calculate the R-squared value for your line above.*

Formula:  
$$ R^2 = 1 - \frac{\sum_i (y_i - m x_i - c)^2}{\sum_i (y_i - \bar{y})^2} $$

```python
rsq = 1.0 - (np.sum((d - m * w - c)**2)/np.sum((d - d_avg)**2))
print("The R-squared value is %6.4f" % rsq)
```

# 10. Use gradient descent  
*Use gradient descent to approximate the best fit line for the petal length and petal width setosa values. Compare the outputs to your calculations above.*  
Formula:  
$$
\begin{align}
Cost(m, c)                       &= \sum_i (y_i - mx_i - c)^2      \\[1cm]
\frac{\partial Cost}{\partial m} &= \sum 2(y_i - m x_i -c)(-x_i) \\
                                 &= -2 \sum x_i (y_i - m x_i -c) \\[0.5cm]
\frac{\partial Cost}{\partial c} & = \sum 2(y_i - m x_i -c)(-1) \\
                                 & = -2 \sum (y_i - m x_i -c)   \\
\end{align}
$$

```python
#partial derivatives
def grad_m(x, y, m, c):
  return -2.0 * np.sum(x * (y - m * x - c))

def grad_c(x, y, m , c):
  return -2.0 * np.sum(y - m * x - c)

#setting learning rate
eta = 0.0001
m, c = 1.0, 1.0
change = True

#iterating, with each iteration m and c get closer to their true value
while change:
  mnew = m - eta * grad_m(setosaLength, setosaWidth, m, c)
  cnew = c - eta * grad_c(setosaLength, setosaWidth, m, c)
  if m == mnew and c == cnew:
    change = False
  else:
    m, c = mnew, cnew
    
print("m: %20.16f  c: %20.16f" % (m, c))    
```

```python
#setting the figure size, title and labels
pl.rcParams['figure.figsize'] = (16, 8)
pl.title("Gradient Descent", fontsize=20)
pl.xlabel("Petal Length", fontsize = 12)
pl.ylabel("Petal Width", fontsize = 12)

#plotting the data and the line
pl.scatter(setosaLength, setosaWidth, marker='x', label='Data')
pl.plot(setosaLength, y, 'g', label='Best Fit Line')
#displaying the legend and the diagram
pl.legend()
pl.show()
```
